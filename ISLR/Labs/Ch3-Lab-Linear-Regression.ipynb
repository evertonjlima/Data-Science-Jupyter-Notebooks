## Linear Regression on Boston dataset

### Loading libraries

```r
library(MASS)
library(ISLR)
```

```r
?Boston
```

```r
lm.fit <- lm(medv~lstat,data=Boston)
```

```r
lm.fit
summary(lm.fit)
```

```r
names(lm.fit)
```

```r
coef(lm.fit)
```

```r
confint(lm.fit)
```

```r
predict(lm.fit,data.frame(lstat=(c(5,10,15))),
                                 interval="confidence")

predict(lm.fit,data.frame(lstat=(c(5,10,15))),
                                 interval="prediction")
```

```r
attach(Boston)
plot(lstat,medv)
abline(lm.fit)
```

```r
par(mfrow = c(2,2))
plot(lm.fit)
```

```r
plot(hatvalues(lm.fit),col=ifelse(hatvalues(lm.fit)==hatvalues(lm.fit)[375],"red","black"))
```

```r
lm.fit <- lm(medv~lstat+age,data=Boston)
summary(lm.fit)
```

```r
lm.fit=lm(medv~.,data=Boston)
summary(lm.fit)
```

```r
lm.fit1=lm(medv~.-age,data=Boston)
summary(lm.fit1)
```

```r
summary(lm(medv~lstat*age,data=Boston))
```

```r
lm.fit2=lm(medv~lstat+I(lstat^2))
summary(lm.fit2)
```

```r
lm.fit=lm(medv~lstat)
anova(lm.fit,lm.fit2)
```

```r
par(mfrow=c(2,2))
plot(lm.fit2)
```

```r
lm.fit5=lm(medv~poly(lstat,5))
summary(lm.fit5)
```

```r
lm.fit8=lm(medv~poly(lstat,8))
summary(lm.fit8)
```
